## Electricity Bill Calculator

First 100 units → 5 per unit

Next 100 units → 7 per unit

Beyond 200 → 10 per unit

Input Format

Integer units.

Constraints

0 ≤ units ≤ 10000

Output Format

Print total bill as integer.

Sample Input 0

250

Sample Output 0

1700

Sample Input 1

50

Sample Output 1

250

Sample Input 2

150

Sample Output 2

850

```python
x = int(input())

if x<=100:
    ans=x*5

elif x<=200:  
    ans= (100*5) + ((x-100)*7)

else:
    ans=(100*5) + (100*7) + ((x-200)*10)

print(ans)
```
