## Fibonacci Series

Write a program to calculate the Fibonacci series i.e., 0 1 1 2 3 5 8 13 21....., up to the given limit n, 
print the result to the console as shown in the example.

Input Format

read integer input

Constraints

0<=n<=100

Output Format

sum previous two numbers and print in new line

Sample Input 0

5

Sample Output 0

0

1

1

2

3

5

Sample Input 1

4

Sample Output 1

0

1

1

2

3

```python
n = int(input())

a, b = 0, 1

while a <= n:
    
    print(a)
    
    a, b = b, a + b
```
