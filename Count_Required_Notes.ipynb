## Count Required Notes

Given the shopping bill Total Amount. Findout How many 2000, 500, 200, 100, 50, 20, 10, 5, 2, 1 currency is required to pay the bill

Input Format

Read the integer input from the user for Total amount

Constraints

Continue

Output Format

If note is reuired print note and how many notes are reuired otherwise print note reuirement is 0. see the sample test case explnation

Sample Input 0

56738

Sample Output 0

2000: 28

500: 1

200: 1

100: 0

50: 0

20: 1

10: 1

5: 1

2: 1

1: 1

Explanation 0

The total amount is : 56738 2000 notes are required 28 500 notes are required 1 200 notes are required 1 100 notes are required 0 50 notes are required 0 20 notes are required 1 10 notes are required 1 5 rupee coin required 1 2 rupee coin required 1 1 Rupee coin required 1

Sample Input 1

630

Sample Output 1

2000: 0

500: 1

200: 0

100: 1

50: 0

20: 1

10: 1

5: 0

2: 0

1: 0

```python
amount = int(input())

deno = [2000, 500, 200, 100, 50, 20, 10, 5, 2, 1]

for d in deno:
    count = amount // d
    print(f"{d}: {count}")
    amount = amount % d
```
