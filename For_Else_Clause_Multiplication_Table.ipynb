It provides a way to perform additional actions or handle specific conditions after the loop has finished iterating.

Explanation

So we print the table only up to 10 rows.

After that, we print "Limit Exceeded".

```python
x = int(input())
y = int(input())

for i in range(1,y+1):
    
    if i <= 10:
        print(f"{x} * {i} = {x*i}")
    
    else:
        print("rows are limited to 10")
        break
```
