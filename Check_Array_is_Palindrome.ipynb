## Check if Array is a Palindrome

You are given an array containing either integers or characters. Write a program that checks whether the array forms a palindrome.

If the first element equals the last, the second equals the second-last, and so on, then the array is a palindrome.

Return true if it is a palindrome, else return false.

Input Format

The first line contains an integer N ‚Äî the size of the array.

The second line contains N space-separated integers or characters.

Constraints

1‚â§N‚â§10^5 Array elements can be: 

Integers in the range ‚àí10^9‚â§ùê¥[ùëñ]‚â§10^9 

Single lowercase/uppercase characters (a‚Äìz, A‚ÄìZ)

Time complexity expected: O(N)

Space complexity: O(1) (excluding input storage)

Output Format

Print true if the array is a palindrome.

Print false otherwise.

Sample Input 0

5

1 2 3 2 1

Sample Output 0

true

Sample Input 1

4

1 2 2 1

Sample Output 1

true

Sample Input 2

6

x y z y x w

Sample Output 2

false

```python
n = int(input())
arr = input().split()

left = 0
right = n - 1

palindrome = True

while left < right:
    if arr[left] != arr[right]:
        palindrome = False
        break
    left += 1
    right -= 1

print(str(palindrome).lower())
```
