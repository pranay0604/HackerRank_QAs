## Find Pair With Minimum Difference

Given a sorted integer array of size N, identify the two adjacent elements whose absolute difference is minimum.

Since the array is already sorted, the minimum difference will always be found between adjacent elements. 
You must print the pair (the two numbers) with the smallest difference.

Return the pair having the minimum difference. If multiple pairs have the same minimum difference, return the first pair from the left.

Input Format

The first line contains an integer N — the size of the array.

The second line contains N space-separated integers (sorted in non-decreasing order).

Constraints

2≤N≤10^5 10^9≤A[i]≤10^9

Array is already sorted

Expected time complexity: O(N)

Expected space complexity: O(1)

Output Format

Print the two integers of the pair that has the minimum difference, separated by a space.

Sample Input 0

5

1 3 4 8 10

Sample Output 0

3 4

Sample Input 1

4

2 5 9 12

Sample Output 1

2 5

Sample Input 2

6

-10 -5 -3 -1 4 8

Sample Output 2

-5 -3

```python
n = int(input())
arr = list(map(int, input().split()))

min_diff = float('inf')
pair = (arr[0], arr[1])

for i in range(n - 1):
    
    diff = abs(arr[i + 1] - arr[i])
    
    if diff < min_diff:
        min_diff = diff
        pair = (arr[i], arr[i + 1])

print(pair[0], pair[1])
```
